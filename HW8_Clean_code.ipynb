````python
import pandas as pd
import numpy as np

np.random.seed(0)

products = ['apple', 'banana', 'cherry', 'dates', 'elderberry']
categories = ['fruit', 'fruit', 'fruit', 'fruit', 'fruit']

N_ROWS = 10**6
df = pd.DataFrame({
    'product_name': np.random.choice(products, N_ROWS),
    'product_category': np.random.choice(categories, N_ROWS),
    'product_price': np.random.uniform(1, 10, N_ROWS),
    'transaction_date': pd.date_range(start='1/1/2000', periods=N_ROWS, freq='h')
})
````

````python
df_new = df.copy()

for col in df_new.columns[0:2]:
    df_new[col] =  pd.Categorical(df_new[col])
````

````python
dif_memory = df.memory_usage(index=False).sum() / df_new.memory_usage(index=False).sum()
````

````python
print(f"Новий датафрейм займає у {round(dif_memory,1)} разів менше місця в пам'яті.")
````

````python
def calculate_variance(number_list):
    """
    Calculate the variance of a list of numbers.
    
    Variance measures how much the values in a list vary from the mean.
    It's calculated as the average of the squared differences between each value and the mean.
    
    Args:
        number_list (list): List of numeric values.
        
    Returns:
        float: Variance of the input list.
    """
    # Calculate the sum of all values in the list
    sum_list = sum(number_list)
    
    # Calculate the mean (average) of the list
    mean = sum_list / len(number_list)
    
    # Calculate the sum of squared differences from the mean
    sum_squared_diff = sum((n - mean) ** 2 for n in number_list)
    
    # Calculate the variance using the corrected formula
    variance = sum_squared_diff / len(number_list)
    
    return variance
````
